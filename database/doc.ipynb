```python
import pandas as pd

# 替换为你实际的 GWAS 文件路径
input_file = "/Users/caicai/Downloads/gwas_catalog_v1.0.2-associations_e113_r2025-04-14.tsv"
output_file = "gwas.csv"

# 加载数据
df = pd.read_csv(input_file, sep='\t', low_memory=False)

# 保留关键字段
columns_to_keep = [
    "SNPS",
    "DISEASE/TRAIT",
    "MAPPED_TRAIT_URI",
    "P-VALUE",
    "OR or BETA",
    "MAPPED_GENE"
]

filtered_df = df[columns_to_keep].dropna(subset=["SNPS", "P-VALUE"])

# 保存为新的 CSV 文件
filtered_df.to_csv(output_file, index=False)

print(f"Filtered GWAS data saved to: {output_file}")
```
